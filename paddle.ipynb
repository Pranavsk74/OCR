```python
from paddleocr import PaddleOCR
import editdistance

# Initialize PaddleOCR with your desired configuration
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # You can change the language if needed

# Ground truth text and OCR output text
ground_truth = "70 UPCYCLED BENCHES IN BKC A at M F t L-R: Dr Sanjay Mukherjee, Mike Hankey and Shishir Joshi.HT Correspondent htmumbai@hindustantimes.com MUMBAI: Mike Hankey, the USConsul General, on Monday inaugurated 70 upcycled benches installed in Bandra Kurla Complex area for visa applicants who queue up outside the US Consulate and Indian Regional Passport Office (IRPO). The benches made of recycled plastic, created by Project Mum- bai, also aims to spread to mes- sage of sustainable living.The project was launched in March. The benches were inau- gurated in collaboration with the Mumbai Metropolitan Region Development Authority and other stakeholders."  # The correct text
ocr_output = "70UPCYCLED BENCHES IN BKO A a L-R:Dr Sanjay Mukherjee,Mike Hankey and Shishir Joshi.HT Correspondent htmumbai@hindustantimes.com MUMBAl:Mike Hankey.the US Consul General. on Monday inaugurated 70 upcycled benches installed in Bandra Kurla Complex area for visa applicants who queue up outside the US Consulate and Indian Regional Passport Office (IRPO).The benches made ofrecycled plastic,created by Project Mum-bai, also aims to spread to mes- sage ofsustainable living The project was launched in March.The benches were inau gurated in collaboration with the Mumbai Metropolitan Region Development Authority and otherstakeholders"     # The text returned by PaddleOCR

# Function to calculate Character Error Rate (CER)
def calculate_cer(ground_truth, ocr_output):
    return editdistance.eval(ground_truth, ocr_output) / float(len(ground_truth))

# Calculate CER
cer = calculate_cer(ground_truth, ocr_output)

# Calculate accuracy
accuracy = 1 - cer

print(f"Character Error Rate (CER): {cer:.4f}")
print(f"Character Accuracy: {accuracy:.4f}")
```
